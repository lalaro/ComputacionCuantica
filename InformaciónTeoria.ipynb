## Escenario de ejemplo: comunicación entre Alice y Bob

Alice quiere enviarle un mensaje a Bob, que consta de uno de cuatro símbolos: A,
B, C o D. 

Su codificación y transmisión 
se puede optimizar en función de la distribución de probabilidad de estos símbolos.

### Distribuciones de probabilidad

1. **Distribución uniforme**: Cada símbolo (A, B, C, D) tiene la misma probabilidad de $ \frac{1}{4} $.
2. **Distribución constante**: Sólo un símbolo, P(A)=1.
3. **Distribución intermedia**: una distribución no uniforme donde ningún mensaje domina, por ejemplo, P(A)=0,5, P(B)=0,25, P(C)=0,125, P(D)=0,125 .

## Programas Python para Cálculo y Visualización


```python
import numpy as np
import matplotlib.pyplot as plt

uniform_dist = np.array([0.25, 0.25, 0.25, 0.25])
constant_dist = np.array([1, 0, 0, 0])
mixed_dist = np.array([5, 0, 1, 1])

def plot_distribution(probabilidades, título):
    etiquetas = ['A', 'B', 'C', 'D']
    plt.bar(etiquetas, probabilidades, color='skyblue')  # Changed color to 'skyblue'
    plt.title(título)
    plt.xlabel('Símbolos')
    plt.ylabel('Probabilidad')
    plt.ylim(0, 1)
    plt.show()

plot_distribution(uniform_dist, 'Distribución uniforme')
plot_distribution(constant_dist, 'Distribución de contactos')
plot_distribution(mixed_dist, 'Distribución mixta')
```

# Explicación de la teoría de la información de Shannon

La teoría de la información de Shannon, introducida por Claude Shannon en la década de 1940, es fundamental para comprender la transmisión, el procesamiento y el almacenamiento de información. Cuantifica la cantidad de incertidumbre en un mensaje y forma la base de los sistemas de comunicación modernos. A continuación, veremos Explore esta teoría a través de un ejemplo que involucra a Alice y Bob comunicándose usando un alfabeto de cuatro símbolos: A, B, C y D.

## Conceptos clave

### Entropía (H)

La entropía mide la incertidumbre o aleatoriedad de un sistema. En teoría de la información, cuantifica el valor esperado de la información contenida en un mensaje.

La entropía de Shannon de una fuente con distribución de probabilidad $\\{p_i\\}$ es:

$$
H_S= -\sum_{i=1}^{n} p_i \log_2 p_i = \sum_{i=1}^{n} p_i \log_2 \frac{1}{p_i}
$$

Donde $p_i$ es la probabilidad de que ocurra el mensaje $x_i$. La unista de H son bits.

### Contenido de información

El contenido de la información se relaciona con la cantidad de información proporcionada por un evento específico, que depende inversamente de la probabilidad de ese evento. Los eventos menos probables contienen más información.

# Programa Python para entropía

```python
import numpy as np
import matplotlib.pyplot as plt

def entropy(probabilities):
    value = 0
    for pi in probabilities:
        if (pi > 0):
            value = value - (pi * (np.log2(pi)))
    return value 


entropy_uniform = entropy (uniform_dist)
entropy_constant = entropy (constant_dist)
entropy_mixed = entropy (mixed_dist)

# Plotting the distributions
plot_distribution(uniform_dist, 'Uniform Distribution (Entropy={:.2f})'.format(entropy_uniform))
plot_distribution(constant_dist, 'Constant Distribution (Entropy={:.2f})'.format(entropy_constant))
plot_distribution(mixed_dist, 'Mixed Distribution (Entropy={:.2f})'.format(entropy_mixed))
```

## La entropía de Von Neumann: Revelando los Misterios del Universo Cuántico

En el ámbito de la física cuántica, la entropía de Von Neumann emerge como una herramienta esencial para entender la incertidumbre y la información contenida en los sistemas cuánticos. A diferencia de la teoría clásica de la información, que se basa en variables discretas, la información cuántica explora un mundo de estados descritos por vectores complejos en un espacio matemático llamado espacio de Hilbert.

### ¿Qué significa la entropía de Von Neumann?

La entropía de Von Neumann, propuesta por el físico y matemático John von Neumann, es una medida de la incertidumbre sobre el estado cuántico de un sistema. Matemáticamente, se expresa como: S(ρ) = -Tr(ρ * log(ρ)) donde:
- ρ es la matriz densidad que describe el estado cuántico.
- Tr es la traza de la matriz.
- log es el logaritmo en base 2.

### Aspectos Clave de la Entropía de Von Neumann:

Comparación con la Entropía de Shannon: La entropía de Von Neumann está relacionada con la entropía de Shannon de la teoría clásica de la información, pero se aplica específicamente a los estados cuánticos.

a. Rango de Valores: La entropía de Von Neumann toma valores entre 0 y log(d), donde d es la dimensión del espacio de Hilbert del sistema.

b. Estado Puro: La entropía de Von Neumann es cero (S(ρ) = 0) solo cuando el estado cuántico es puro, es decir, descrito por un único vector de estado.

c. Estado Máximamente Mezclado: Cuando el sistema está en un estado máximamente mezclado (representado por una matriz densidad proporcional a la matriz identidad), la entropía de Von Neumann alcanza su valor máximo, log(d).

### Aplicaciones de la Entropía de Von Neumann:

1. *Medición de la Información Cuántica:* La entropía de Von Neumann es fundamental en la teoría de la información cuántica, ya que permite cuantificar la información codificada en un estado cuántico.

2. *Indicador de Pureza:* También se usa para determinar la pureza de un estado cuántico. Un estado puro tiene S(ρ) = 0, mientras que un estado mezclado tiene una entropía cercana a log(d).

3. *Caracterización del Entrelazamiento:* Se utiliza para caracterizar la cantidad de entrelazamiento en un sistema compuesto, una propiedad importante en la mecánica cuántica.

**Conceptos Relacionados:** La entropía de Von Neumann está relacionada con otros conceptos en la teoría de la información cuántica, como la compresión de datos cuánticos y el teorema de codificación cuántica.

En resumen, la entropía de Von Neumann se construye como una herramienta indispensable para comprender y cuantificar la información en el fascinante mundo de la mecánica cuántica. Su aplicación abarca desde la medición de la incertidumbre y la pureza hasta la caracterización del entrelazamiento y el desarrollo de tecnologías cuánticas novedosas.

### Entropía y positividad de Shannon

La entropía de Shannon, denotada por H(p), es una medida de incertidumbre asociada con una distribución de probabilidad p = {p₁,..., pn} donde pi representa la probabilidad de que ocurra el evento i. Se calcula mediante la siguiente fórmula:

H(p) = - Σ (pi * log2(pi))

Aquí, la suma itera sobre todos los eventos posibles, y log2(pi) es el logaritmo de pi en base 2. Curiosamente, la entropía de Shannon satisface una propiedad en la que siempre es no negativa (positiva o cero). Demostremos esto matemáticamente.

Por tanto, la entropía de Shannon (H(p)) siempre es no negativa. Es igual a cero sólo cuando no hay incertidumbre (toda la masa de probabilidad concentrada en un solo evento). En todos los demás casos, donde existe cierto grado de incertidumbre sobre qué evento ocurrirá, la entropía será positiva.

## 4.1 Estados cuánticos como se describe en la sección.

```python
import numpy as np

def estado_base(n_qubits):
  """
  Crea un estado base |0>^n de n qubits.

  Args:
    n_qubits: Número de qubits.

  Returns:
    Estado base |0>^n como vector numpy.
  """
  return np.array([1] + [0] * (2 ** n_qubits - 1))

def estado_superposicion_igual(n_qubits):
  """
  Crea un estado de superposición uniforme |1/sqrt(2)>^n de n qubits.

  Args:
    n_qubits: Número de qubits.

  Returns:
    Estado de superposición uniforme |1/sqrt(2)>^n como vector numpy.
  """
  # Se define la variable n dentro de la función
  n = n_qubits
  superposicion = np.array([1 / np.sqrt(2**n)] * 2**n)
  return superposicion


def estado_arbitrario(probabilidades):
  """
  Crea un estado cuántico arbitrario a partir de una lista de probabilidades.

  Args:
    probabilidades: Lista de probabilidades para cada estado base.

  Returns:
    Estado cuántico arbitrario como vector numpy.
  """
  return np.array(probabilidades)

# Ejemplo de uso
n_qubits = 2

estado_base_dos_qubits = estado_base(n_qubits)
estado_superposicion_igual_dos_qubits = estado_superposicion_igual(n_qubits)

print("Estado base de 2 qubits:", estado_base_dos_qubits)
print("Estado superposición uniforme de 2 qubits:", estado_superposicion_igual_dos_qubits)
```

```python
def matriz_densidad(estado):
  """
  Calcula la matriz de densidad para un estado cuántico puro.

  Args:
    estado: Vector de estado puro como vector numpy.

  Returns:
    Matriz de densidad correspondiente al estado puro.
  """
  estado_externo = np.outer(estado, estado.conj())
  return estado_externo

# Ejemplo de uso
estado_puro = estado_superposicion_igual_dos_qubits
matriz_densidad_estado_puro = matriz_densidad(estado_puro)

print("Matriz de densidad para estado puro:", matriz_densidad_estado_puro)
```

## 4.3. Calcula la entropía de Von Neumann para las matrices de densidad construidas.

```python
import scipy.stats as stats

def entropia_von_neumann(matriz_densidad):
  """
  Calcula la entropía de Von Neumann para una matriz de densidad.

  Args:
    matriz_densidad: Matriz de densidad como matriz numpy.

  Returns:
    Entropía de Von Neumann de la matriz de densidad.
  """
  valores_propios = stats.entropy(matriz_densidad.ravel())
  return -np.sum(valores_propios * np.log2(valores_propios))

# Ejemplo de uso
entropia_estado_puro = entropia_von_neumann(matriz_densidad_estado_puro)
print("Entropía de Von Neumann para estado puro:", entropia_estado_puro)
```

## 4.4 Simule un escenario de comunicación cuántica simple entre dos partes, destacando el papel de la entropía de Von Neumann.

```python
import numpy as np
import scipy.stats as stats

n_qubits = 1  # Número de qubits
p_perdida = 0.2  # Probabilidad de pérdida en el canal de pérdida
tasa_dephasing = 0.1  # Tasa de decoherencia en el canal de dephasing


def estado_base(n_qubits):
  """
  Crea un estado base |0>^n de n qubits.

  Args:
    n_qubits: Número de qubits.

  Returns:
    Estado base |0>^n como vector numpy.
  """
  return np.array([1] + [0] * (2 ** n_qubits - 1))

def estado_superposicion_igual(n_qubits):
  """
  Crea un estado de superposición uniforme |1/sqrt(2)>^n de n qubits.

  Args:
    n_qubits: Número de qubits.

  Returns:
    Estado de superposición uniforme |1/sqrt(2)>^n como vector numpy.
  """
  return np.array([1 / np.sqrt(2**n)] * 2**n)

def estado_arbitrario(probabilidades):
  """
  Crea un estado cuántico arbitrario a partir de una lista de probabilidades.

  Args:
    probabilidades: Lista de probabilidades para cada estado base.

  Returns:
    Estado cuántico arbitrario como vector numpy.
  """
  return np.array(probabilidades)

def matriz_densidad(estado):
  """
  Calcula la matriz de densidad para un estado cuántico puro.

  Args:
    estado: Vector de estado puro como vector numpy.

  Returns:
    Matriz de densidad correspondiente al estado puro.
  """
  estado_externo = np.outer(estado, estado.conj())
  return estado_externo

def entropia_von_neumann(matriz_densidad):
  """
  Calcula la entropía de Von Neumann para una matriz de densidad.

  Args:
    matriz_densidad: Matriz de densidad como matriz numpy.

  Returns:
    Entropía de Von Neumann de la matriz de densidad.
  """
  valores_propios = stats.entropy(matriz_densidad.ravel())
  return -np.sum(valores_propios * np.log2(valores_propios))

# Definición de canales cuánticos (ejemplos)
def canal_perdida(estado, p_perdida):
  """
  Canal de pérdida que elimina una fracción p_perdida de la información del qubit.

  Args:
    estado: Estado cuántico como vector numpy.
    p_perdida: Probabilidad de pérdida de información.

  Returns:
    Estado mixto resultante del canal de pérdida.
  """
  n_qubits = len(estado)  # **Definir n_qubits antes de llamar a la función**
  estado_conservado = (1 - p_perdida) * estado
  estado_perdido = np.zeros_like(estado)
  probabilidad_conservado = 1 - p_perdida
  probabilidad_perdido = p_perdida

  matriz_densidad_mixta = probabilidad_conservado * np.outer(estado_conservado, estado_conservado.conj()) + \
                         probabilidad_perdido * np.outer(estado_perdido, estado_perdido.conj())
  return matriz_densidad_mixta
```
